Написать приложение и функцию, которые собирают основные новости с сайта на выбор dzen.ru, lenta.ru, mail.ru . Для парсинга использовать XPath
Структура данных должна содержать:
* название источника
* наименование новости
* ссылку на новость
* дата публикации

минимум один сайт максимум все

```python
from pprint import pprint
from lxml import html
import requests
import pandas as pd
```

#### dzen.ru

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
url_dzen = 'https://dzen.ru/news/'
```

```python
headers = {
    'User-Agent': user_agent,
}

params = {
    'sso_failed': '',
}
```

```python
response_dzen = requests.get(url_dzen, headers=headers, params=params)
content_dom_dzen = html.fromstring(response_dzen.text)

print(response_dzen.url)
```

```python
news_container_dzen = content_dom_dzen.xpath("//div[contains(@class,'mg-card ')]")

news_dzen = []
for new_container_dzen in news_container_dzen:
    new_source_dzen = new_container_dzen.xpath(".//a[@class='mg-card__source-link']/text()")[0]
    new_title_dzen = new_container_dzen.xpath(".//h2/a/text()")[0].replace('\xa0', ' ')
    new_link_dzen = new_container_dzen.xpath(".//h2/a/@href")[0]
    new_date_dzen = new_container_dzen.xpath(".//span[@class='mg-card-source__time']/text()")[0]
    news_dict_dzen = {
            'new_source_dzen': new_source_dzen,
            'new_title_dzen': new_title_dzen,
            'new_link_dzen': new_link_dzen,
            'new_date_dzen': new_date_dzen,
        }
    news_dzen.append(news_dict_dzen)
    
news_dzen
```

#### lenta.ru

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
url_lenta = 'https://lenta.ru/parts/news/'
```

```python
headers = {
    'User-Agent': user_agent,
}

params = {
    'sso_failed': '',
}

response_lenta = requests.get(url_lenta, headers=headers, params=params)
content_dom_lenta = html.fromstring(response_lenta.text)

print(response_lenta.url)
```

```python
news_container_lenta = content_dom_lenta.xpath("//ul[contains(@class,'parts-page__body ')]/li")

len(news_container_lenta)
```

```python
news_container_lenta = content_dom_lenta.xpath("//ul[contains(@class,'parts-page__body ')]/li")

news_lenta = []
for new_container_lenta in news_container_lenta:
    new_source_lenta = new_container_lenta.xpath(".//div/span/text()")
    if len(new_source_lenta) == 1:
        new_source_lenta = new_source_lenta[0]
    else:
        new_source_lenta = 'Error data'
    
    new_title_lenta = new_container_lenta.xpath(".//h3/text()")
    if len(new_title_lenta) == 1:
        new_title_lenta = new_title_lenta[0]
    else:
        new_source_lenta = 'Error data'
    
    new_link_lenta = new_container_lenta.xpath("./a/@href")
    if len(new_link_lenta) == 1:
        new_link_lenta = new_link_lenta[0]
    else:
        new_link_lenta = 'Error data'
    
    new_date_lenta = new_container_lenta.xpath(".//time/text()")
    if len(new_date_lenta) == 1:
        new_date_lenta = new_date_lenta[0]
    else:
        new_date_lenta = 'Error data'
    
    news_dict_lenta = {
            'new_source_lenta': new_source_lenta,
            'new_title_lenta': new_title_lenta,
            'new_link_lenta': new_link_lenta,
            'new_date_lenta': new_date_lenta,
        }
    news_lenta.append(news_dict_lenta)
    
news_lenta
```

#### mail.ru

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
url_mail = 'https://news.mail.ru/'
```

```python
headers = {
    'User-Agent': user_agent,
}

params = {
    'sso_failed': '',
}

response_mail = requests.get(url_mail, headers=headers, params=params)
content_dom_mail = html.fromstring(response_mail.text)

print(response_mail.url)
```

```python
news_container_mail = content_dom_mail.xpath("//td/div[contains(@class,'daynews__item')]")

len(news_container_mail)
```

```python
news_mail = []

for new_container_mail in news_container_mail:
    
    new_title_mail = new_container_mail.xpath(".//span[contains(@class,'photo__title ')]/text()")[0].replace('\xa0', ' ')
        
    new_link_mail = new_container_mail.xpath(".//a/@href")[0]
    
    news_dict_mail = {
            'new_title_mail': new_title_mail,
            'new_link_mail': new_link_mail        
        }
    news_mail.append(news_dict_mail)
    
news_mail
```
